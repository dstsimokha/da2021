```r
library(sjPlot)
library(ggplot2)
```

```r
options(repr.plot.width = 20, repr.plot.height = 15)
theme_set(theme_bw(base_size = 30))
update_geom_defaults("smooth", list(size = 5))
```

```r
load('LM3.RData')
```

```r
model0 = lm(sat ~ age1 + sex + eduf + sex + H_URBRURAL + eduf + income, data = ruswvs7)
summary(model0)
```

```r
model1 = lm(sat~age1+sex*eduf+H_URBRURAL+income, data = ruswvs7)
summary(model1)
```

```r
plot_model(model1, line.size = 5,
           type = "pred", terms=c("eduf", "sex"), 
           title = "", axis.title= c("education level","life satisfaction"))
```

```r
model2 = lm(sat~age1+sex*eduf+H_URBRURAL+eduf*income, data = ruswvs7)
summary(model2)
```

```r
plot_model(model2, line.size = 5,
           type = "pred", terms=c("eduf", "income"), 
           title = "", axis.title= c("education level","life satisfaction")) 
```

```r
model2.1 = lm(sat~age1+sex*edu+H_URBRURAL+edu*income, data = ruswvs7)
summary(model2.1)
```

```r
plot_model(model2.1, line.size = 5,
           type = "pred", terms=c("income", "edu"), 
           title = "", axis.title= c("incomelevel", "life satisfaction"))
```

```r
plot_model(model2.1, line.size = 5, dot.size = 15,
           type = "pred", terms=c("sex", "edu"), 
           title = "", axis.title= c("gender", "life satisfaction"))
```

# Transformation

## Quadratic

```r
model3 = lm(sat~age1+I(age1^2)+sex*edu+H_URBRURAL+edu*income, data = ruswvs7)
summary(model3)
```

```r
ggplot(ruswvs7, aes(x = age1, y = sat)) + 
stat_smooth(se = F, method = 'lm', formula = y~x+I(x^2))
```

## Scaling

```r
model4 = lm(data = ruswvs7, 
            sat ~ scale(age1) + I(scale(age1)^2) + sex*edu + H_URBRURAL + edu*income)
summary(model4)
```

```r
ggplot(ruswvs7, aes(x = scale(age1), y = sat)) + 
stat_smooth(se = F, method = 'lm', formula=y~x+I(x^2))
```

## Log

```r
model5 = lm(sat ~ age1 + sex + edu + H_URBRURAL + log(income), data = ruswvs7)
summary(model5)
```
